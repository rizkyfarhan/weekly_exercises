## Soal 2

```python
data2 = pd.read_csv('datasoal1.csv')
```

```python
data2.head(3)
```

```python
nt = list()
for i in range(data2.shape[0]):
    nt.append(data2.drop(['Hasil Jual Tanam','Hari Ke'],axis=1).loc[i].median())
data2['Nilai Tengah'] = nt
```

```python
data2.head(3)
```

```python
pj = list()
for i in range(data2.shape[0]):
    if i % 2 == 0 and i % 3 != 0 and i%5 != 0:
        pj.append('MSY')
    elif i % 3 == 0 and i % 2 == 0 and i % 5 != 0:
        pj.append('TRN')
    elif i % 3 == 0 and i % 2 != 0 and i % 5 != 0:
        pj.append('TRN')
    elif i % 5 == 0 and i % 3 == 0 and i % 2 == 0:
        pj.append('NSR')
    elif i % 5 == 0 and i % 3 != 0 and i % 2 == 0:
        pj.append('NSR')
    elif i % 5 == 0 and i % 3 == 0 and i % 2 != 0:
        pj.append('NSR')
    elif i % 5 == 0 and i % 3 != 0 and i % 2 != 0:
        pj.append('NSR')
    else:
        pj.append('AKT')
data2['Penanggung Jawab'] = pj
```

```python
data2.head(3)
```

```python
data2.describe()
```

```python
data2['Penanggung Jawab'].value_counts()
```

```python
max = data2.groupby('Penanggung Jawab').max()
max
```

```python
min = data2.groupby('Penanggung Jawab').min()
min
```

## Soal 4

```python
oz = pd.read_csv('ozone.csv')
```

```python
avg = list()
for i in range(oz.shape[0]):
    avg.append(oz.loc[i].mean())
oz['Rata-Rata'] = avg
```

```python
ktg = list()
for i in range(oz.shape[0]):
    if oz['Wind_speed'][i]<=5:
        ktg.append('Rendah')
    elif oz['Wind_speed'][i]>5 and oz['Wind_speed'][i]<=10:
        ktg.append('Sedang')
    elif oz['Wind_speed'][i]>10:
        ktg.append('Tinggi')
oz['Kategori'] = ktg
```

```python
oz
```

```python
bykat = oz.groupby('Kategori')
```

```python
bykat.max()
```

```python
bykat.min()
```

```python
oz.mean()
```

```python
bykat.describe()
```

```python
col = ['r','g','b','y']
idx = np.arange(oz.shape[0])
for i in range(14):
    plt.subplot(7,2,i+1)
    plt.plot(idx,oz[oz.columns[i]],np.random.choice(col))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

idx = np.arange(oz.shape[0])
for i in range(14):
    ax.plot(idx,oz[oz.columns[i]],label=oz.columns[i])
ax.legend()
```

```python
mpg
```

```python
mpg.head(5)
```

```python
mpg.tail(10)
```

```python
mpg.describe()
```

```python
mpg.isnull().any()
```

```python
mpg['horsepower'].fillna(value=mpg['horsepower'].mean(),inplace=True)
```

```python
mpg.isnull().any()
```

```python
hp = list()
for i in range(mpg.shape[0]):
    if mpg['horsepower'][i]<=105:
        hp.append('lower ' + mpg['origin'][i])
    elif mpg['horsepower'][i]>105:
        hp.append('upper ' + mpg['origin'][i])
mpg['category'] = hp
```

```python
mpg
```

```python
byor = mpg.groupby('origin')
byor.describe()
```

```python
bycat = mpg.groupby('category')
bycat.describe()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

idx = np.arange(mpg.shape[0])
for i in range(7):
    ax.plot(idx,mpg[mpg.columns[i]],label=mpg.columns[i])
ax.set_title('Grafik Gabungan')
ax.set_xlabel('Indeks')
ax.set_ylabel('Nilai')
ax.legend()
```
