## Soal 1

```python
idx = list()
for i in range(3):
    for j in range(1,31):
        idx.append(j)
data1 = pd.DataFrame(index=idx,columns='A B C D'.split())
```

```python
data1
```

```python
satu = list()
for i in idx:
    if i * 3 > 50:
        satu.append(i)
    else:
        satu.append(i*3)
data1['A'] = satu
```

```python
data1['B'] = np.random.randint(1,11,90)
```

```python
data1['C'] = data1['B'] ** 2
```

```python
empat = list()
sel = list()
sel = data1['C'] - data1['B']
for i in sel:
    if i<10:
        empat.append('Nilai Rendah')
    elif i>15:
        empat.append('Nilai Tinggi')
    else:
        empat.append('Nilai Sedang')
data1['D'] = empat
```

```python
data1
```

```python
sns.countplot(data1['D'])
```

```python
sns.boxplot(x='D',y='A',data=data1,palette='rainbow')
```

```python
sns.jointplot(x='C',y='B',data=data1,kind='scatter')
```

```python
digi = pd.read_csv('digimon.csv')
digi.head()
```

```python
d2 = pd.DataFrame(index=np.arange(0,249),columns='Spd HP Def Stg'.split())
```

```python
d2['Spd'] = digi['Lv50 Spd']
```

```python
d2['HP'] = digi['Lv 50 HP'] // 2
```

```python
d2['Def'] = digi['Lv50 Def'] * 3
```

```python
d2['Stg'] = digi['Stage']
```

```python
rank = list()
for i in range(len(d2.index)):
    if d2['Spd'][i] > 130:
        rank.append('Fast')
    elif d2['Spd'][i] < 90:
        rank.append('Slow')
    else:
        rank.append('Normal')
d2['Rnk'] = rank
```

```python
d2.describe()
```

```python
sns.countplot(d2['Rnk'])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1.5,0.7])
ax.plot(np.arange(len(d2.index)),d2['Spd'],label='Spd')
ax.plot(np.arange(len(d2.index)),d2['HP'],label='HP')
ax.plot(np.arange(len(d2.index)),d2['Def'],label='Def')
ax.set_title('Attributes')
ax.set_xlabel('Number')
ax.set_ylabel('Value')
ax.legend()
```

## Soal 4

```python
ins = pd.read_csv('insurance.csv')
```

```python
ins.head()
```

```python
d4 = pd.DataFrame(index=np.arange(0,50),columns='bmi age children charges sex smoker region'.split())
```

```python
awal = 0
akhir = 1
bm = list()
for i in range(len(ins.index)//2):
    rt = (ins['bmi'][awal] + ins['bmi'][akhir])/2
    bm.append(rt)
    awal += 2
    akhir += 2
d4['bmi'] = bm
```

```python
i = 0
ag = list()
while i < (len(ins)-2):
    rt = rt = (ins['bmi'][i] + ins['bmi'][i+2])/2
    ag.append(rt)
    if i % 2 == 0:
        i += 1
    else:
        i+= 3
d4['age'] = ag
```

```python
d4['children'] = (ins['children'][0:50])*2
```

```python
d4['charges'] = (ins['charges'][0:50])*2
```

```python
d4['sex'] = ins['sex']
d4['smoker'] = ins['smoker']
d4['region'] = ins['region']
```

```python
d4.head()
```

```python
d4.describe()
```

```python
plt.figure(figsize=(14,3))
plt.subplot(1,3,1)
sns.countplot(x='sex',data=d4)
plt.subplot(1,3,2)
sns.countplot(x='smoker',data=d4)
plt.subplot(1,3,3)
sns.countplot(x='region',data=d4)
```

```python
plt.figure(figsize=(14,3))
plt.subplot(1,4,1)
sns.boxplot(x='sex',y='bmi',data=d4)
plt.subplot(1,4,2)
sns.boxplot(x='sex',y='age',data=d4)
plt.subplot(1,4,3)
sns.boxplot(x='sex',y='children',data=d4)
plt.subplot(1,4,4)
sns.boxplot(x='sex',y='charges',data=d4)
```

```python
sns.stripplot(x='sex',y='charges',data=d4,hue='smoker')
```

```python
pg = sns.PairGrid(d4)
pg.map(plt.scatter)
```
